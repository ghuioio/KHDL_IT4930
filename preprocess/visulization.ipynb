```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# phân phối về giá, đang bị mất cân bằng, có cả ô tô giá < 50 triệu
# có ô to giá lại quá cao 499000000000 

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
fig.set_figheight(5)
fig.set_figwidth(20)

axs = [ax1, ax2, ax3, ax4]
sample = df.gia.values / 1000000 # don vi trieu



for n in (range(0, len(axs))):
    axs[n].hist(sample, bins=10 * 10**n, range=(0, 10 * 10**n))
    axs[n].set_title('Trong khoảng từ 0 tới {} triệu'.format(10 * 10**n))

plt.show();
```

```python
# có 3 mẫu xe có giá cao hơn 100000 triệu, ngoại lai, xem lại data về các mẫu nhỏ hơn 50 triệu và lớn hơn 100000 triệu

fig, ax =plt.subplots()

ax.hist(sample, bins=10, range=(100000, 499000))

plt.show();
```

```python
# phân phối xuất xứ

num = np.array([df[df.xuat_xu == 'Lắp ráp trong nước'].count().values[0], df[df.xuat_xu == 'Nhập khẩu'].count().values[0]])

lable = ('Lắp ráp trong nước', 'Nhập khẩu')
y_pos = np.arange(len(lable))

error = np.random.rand(len(lable))

fig, ax = plt.subplots()

hbars = ax.barh(y_pos, num, xerr=error, align='center')
ax.set_yticks(y_pos)

ax.invert_yaxis()
ax.set_yticklabels(lable)

ax.bar_label(hbars)

plt.show()
```

```python
#gias xe cũ và mới, xe mới giá rẻ nhất là 100 triệu?
hyundai = df[df['hang'] == 'Hyundai']

data = [hyundai.gia /1000000, hyundai.km_da_di]
headers = ['gia', 'km_da_di']
df2 = pd.concat(data, axis=1, keys=headers)
df2.plot(kind='scatter', x='km_da_di', y='gia', ylim=(0, 1000), xlim=(0, 100), figsize = (30, 10));
```

```python
data = [df.gia /1000000, df.nam_sx]
headers = ['gia', 'nam_sx']
df3 = pd.concat(data, axis=1, keys=headers)
df3.sort_values(by=['nam_sx'])
df3.plot(kind='scatter', x='nam_sx', y='gia', ylim=(0, 10000), figsize = (30, 5));

# biểu đồ thể hiện ô tô càng mới sản xuất thì giá cao hơn
```

```python
# phân phối nhiên liệu
fig = plt.figure(figsize = (10,10))
ax = fig.subplots()
df.nhien_lieu.value_counts().plot(ax=ax, kind='pie', autopct='%1.1f%%', fontsize=12)
ax.set_title("Cars Count by Engine Fuel Type")
plt.show()
```

```python
var = df.groupby('nhien_lieu').gia.sum() 
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.set_xlabel('Condition Of Car')
ax1.set_ylabel('Increase In price')
ax1.set_title("Condition Vs Price")
var.plot(kind='bar')
```

```python
var = pd.DataFrame(df.groupby(['nhien_lieu'])['gia'].mean().sort_values(ascending = False))
var.plot.bar()
plt.title('Fuel Type vs Average Price')
plt.show()
```
